```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mean = 100
std_dev = 4
size = 100000
```

```python
data = np.random.normal(loc=mean, scale=std_dev, size=size)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=100, kde=True, color='skyblue')
```
